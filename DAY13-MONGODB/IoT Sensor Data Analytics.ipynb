```python
db.sensor_readings.insertMany([
  { timestamp: ISODate("2025-11-24T08:15:00Z"), temperature: 22.5, humidity: 55 },
  { timestamp: ISODate("2025-11-24T08:45:00Z"), temperature: 23.0, humidity: 57 },
  { timestamp: ISODate("2025-11-24T09:15:00Z"), temperature: 22.8, humidity: 56 },
  { timestamp: ISODate("2025-11-24T09:45:00Z"), temperature: 23.2, humidity: 58 },
  { timestamp: ISODate("2025-11-24T10:15:00Z"), temperature: 23.5, humidity: 59 }
])

db.sensor_readings.aggregate([
  {
    $group: {
      _id: { $hour: "$timestamp" }, // Group by hour
      avg_temperature: { $avg: "$temperature" },
      avg_humidity: { $avg: "$humidity" }
    }
  },
  { $sort: { "_id": 1 } }
])

db.sensor_readings.aggregate([
  {
    $match: {
      timestamp: {
        $gte: ISODate("2025-11-24T00:00:00Z"),
        $lt: ISODate("2025-11-25T00:00:00Z")
      }
    }
  },
  {
    $group: {
      _id: null,
      min_temperature: { $min: "$temperature" },
      max_temperature: { $max: "$temperature" },
      min_humidity: { $min: "$humidity" },
      max_humidity: { $max: "$humidity" }
    }
  }
])
```
