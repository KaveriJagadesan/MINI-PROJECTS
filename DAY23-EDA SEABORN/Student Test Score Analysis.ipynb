```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
df = pd.DataFrame({
    "gender": ["Male", "Female"] * 25,
    "study_group": ["A", "B", "C", "A", "B"] * 10,
    "math_score": [65, 70, 82, 90, 76] * 10,
    "reading_score": [60, 68, 80, 92, 74] * 10
})

df.head()
```

Boxplot — Comparing Scores by Gender

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="gender", y="math_score", data=df)
plt.title("Math Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.show()
```

Violin Plot — Distribution of Scores by Study Group

```python
plt.figure(figsize=(8,5))
sns.violinplot(x="study_group", y="math_score", data=df)
plt.title("Distribution of Math Scores by Study Group")
plt.xlabel("Study Group")
plt.ylabel("Math Score")
plt.show()
sns.violinplot(x="study_group", y="reading_score", data=df)
```

Scatter Plot — Relationship Between Two Subjects

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="math_score", y="reading_score", hue="gender", data=df)
plt.title("Math vs Reading Scores")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
```
