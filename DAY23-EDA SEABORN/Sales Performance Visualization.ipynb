```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample sales dataset
data = {
    "date": pd.date_range(start="2024-01-01", periods=12, freq="M"),
    "sales": [12000, 15000, 18000, 22000, 21000, 24000, 26000, 23000, 25000, 28000, 30000, 32000],
    "region": ["North", "South", "East", "West"] * 3,
    "category": ["Electronics", "Clothing", "Groceries"] * 4
}

df = pd.DataFrame(data)

df.head()
```

Line Plot — Sales Trend Over Time

```python
plt.figure(figsize=(8,5))
sns.barplot(x="region", y="sales", data=df)
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

Bar Plot — Sales by Region

```python
plt.figure(figsize=(8,5))
sns.barplot(x="region", y="sales", data=df)
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

Heatmap — Sales by Month and Category

```python
df["month"] = df["date"].dt.month_name()

pivot_table = df.pivot_table(
    values="sales",
    index="month",
    columns="category",
    aggfunc="sum"
)

pivot_table

plt.figure(figsize=(8,6))
sns.heatmap(pivot_table, annot=True, fmt="g", cmap="YlGnBu")
plt.title("Sales by Month and Product Category")
plt.show()
```
