```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate a time index
dates = pd.date_range(start="2025-01-01", periods=15, freq="D")

# Create a dataset with some missing values
data = {
    "date": dates,
    "sales": [200, 220, np.nan, 250, np.nan, 270, 300, np.nan, 320, 340, np.nan, 360, 380, np.nan, 400]
}

df = pd.DataFrame(data)
df.set_index("date", inplace=True)

print("Original Dataset with Missing Values:")
print(df)
```

Visualize Missing Data

```python
plt.figure(figsize=(10,5))
plt.plot(df.index, df['sales'], marker='o', label='Sales')
plt.title("Sales Time-Series with Missing Values")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

Fill Missing Values Using Time-Based Interpolation

```python
# Interpolate using time method
df_interpolated = df.interpolate(method='time')

print("Dataset After Time-Based Interpolation:")
print(df_interpolated)
```

Visualize the Interpolated Data

```python
plt.figure(figsize=(10,5))
plt.plot(df.index, df['sales'], marker='o', linestyle='--', label='Original')
plt.plot(df_interpolated.index, df_interpolated['sales'], marker='o', label='Interpolated')
plt.title("Time-Series Interpolation of Missing Sales Values")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
