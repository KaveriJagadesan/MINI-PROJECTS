```python
import numpy as np
import pandas as pd
data = {
    "age": [25, 30, np.nan, 45, 35, np.nan, 40],
    "salary": [35000, 42000, 50000, np.nan, 58000, 61000, np.nan],
    "city": ["Delhi", "Mumbai", np.nan, "Chennai", "Delhi", np.nan, "Mumbai"]
}

df = pd.DataFrame(data)
df
```

Identify Missing Values

```python
df.isnull().sum()
```

Handle Numerical Missing Values

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution")
plt.show()

# Age is roughly symmetric → fill with mean
df['age'].fillna(df['age'].mean(), inplace=True)

# Salary is skewed → fill with median
df['salary'].fillna(df['salary'].median(), inplace=True)
```

Handle Categorical Missing Values

```python
# Check value counts
df['city'].value_counts()
# Fill missing values with mode
df['city'].fillna(df['city'].mode()[0], inplace=True)
df
```

Final Check

```python
df.isnull().sum()
```
