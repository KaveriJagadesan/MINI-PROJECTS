```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Generate Synthetic Data

```python
np.random.seed(42)  # for reproducibility
# True relationship: y = 3x + 5 + noise
x = np.linspace(0, 10, 50)
noise = np.random.randn(50) * 2
y = 3 * x + 5 + noise
```

 2. Define Mean Squared Error Loss

```python
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

3. Gradient calculations

```python
def compute_gradients(x, y, y_pred):
    n = len(y)
    dm = (-2/n) * np.sum(x * (y - y_pred))   # derivative wrt m
    db = (-2/n) * np.sum(y - y_pred)         # derivative wrt b
    return dm, db
```

4. Gradient Descent Loop

```python
m = np.random.randn()  # random initial slope
b = np.random.randn()  # random initial intercept

learning_rate = 0.01
epochs = 2000

for epoch in range(epochs):
    y_pred = m * x + b              # predictions
    loss = mse_loss(y, y_pred)      # calculate loss
    dm, db = compute_gradients(x, y, y_pred)  # compute gradient

    # update parameters
    m = m - learning_rate * dm
    b = b - learning_rate * db

    # print progress
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.4f}, m = {m:.4f}, b = {b:.4f}")

# Final learned parameters
print("\nFinal model:")
print(f"m = {m:.4f}, b = {b:.4f}")
```

5. Visualize the result

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label="Data Points")
plt.plot(x, m * x + b, label="Best Fit Line", linewidth=2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression Using Gradient Descent")
plt.legend()
plt.grid(True)
plt.show()
```
