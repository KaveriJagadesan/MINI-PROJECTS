```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Define the function f(x) = x^2

```python
def f(x):
    return x**2
```

2. Define its derivative f'(x) = 2x

```python
def f_prime(x):
    return 2 * x
```

3. Create x values

```python
x = np.linspace(-5, 5, 400)
```

4. Calculate y and y'

```python
y = f(x)
y_prime = f_prime(x)
```

5. Plot both curves

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="f(x) = xÂ²")
plt.plot(x, y_prime, label="f'(x) = 2x")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.title("Function and Its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
