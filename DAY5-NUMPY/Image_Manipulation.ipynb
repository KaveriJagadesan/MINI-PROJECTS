```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Load or create a grayscale image as a NumPy array
# Create a 100x100 pixel fake grayscale image with random values between 0.0 and 1.0
fake_image = np.random.rand(100, 100)

# 2. Perform transformations
# Invert the pixel values
# Subtract each pixel from the maximum possible value (1.0 for this float array)
inverted_image = 1.0 - fake_image

# Crop a central portion of the image
# Slice the array to get the central 50x50 pixels (from index 25 to 75)
cropped_image = fake_image[25:75, 25:75]

# 3. Visualization
# Set up the plot for displaying the images
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Image Transformations with NumPy', fontsize=16)

# Display the original image
axes[0].imshow(fake_image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')  # Hide the axes

# Display the inverted image
axes[1].imshow(inverted_image, cmap='gray')
axes[1].set_title('Inverted Image')
axes[1].axis('off')

# Display the cropped image
axes[2].imshow(cropped_image, cmap='gray')
axes[2].set_title('Cropped Image')
axes[2].axis('off')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()

```
