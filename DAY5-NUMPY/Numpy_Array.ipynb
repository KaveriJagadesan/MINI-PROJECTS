```python
import numpy as np

# 1. Create a sample 2D array with shape (5, 4)
# The values will be random integers between 0 (inclusive) and 100 (exclusive).
arr = np.random.randint(100, size=(5, 4))
print("Original 2D Array:")
print(arr)
print("-" * 30)

# 2. Calculations
# Calculate the mean of each column (axis=0)
column_means = arr.mean(axis=0)
print("Mean of each column:")
print(column_means)

# Calculate the maximum value in each row (axis=1)
row_maxes = arr.max(axis=1)
print("\nMaximum value of each row:")
print(row_maxes)

# Find the total sum of all elements in the array
total_sum = arr.sum()
print("\nTotal sum of all elements:")
print(total_sum)
print("-" * 30)

# 3. Indexing

sliced_array = arr[0:2, 2:4]
print("First two rows and last two columns (slicing):")
print(sliced_array)

greater_than_50 = arr[arr > 50]
print("\nElements greater than 50 (boolean indexing):")
print(greater_than_50)
print("-" * 30)

```
