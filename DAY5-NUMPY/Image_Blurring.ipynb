```python
import numpy as np
import matplotlib.pyplot as plt

def simple_blur(image_array):
    
    # Get image dimensions
    rows, cols = image_array.shape

    # Create a new array of zeros to store the blurred image.
    # We use a float dtype to handle division for the average.
    blurred_image = np.zeros((rows, cols), dtype=np.float64)

    # Iterate through each pixel, excluding a 1-pixel border to simplify edge handling.
    for r in range(1, rows - 1):
        for c in range(1, cols - 1):
            # Extract the 3x3 block of neighbor pixels
            neighbors = image_array[r-1:r+2, c-1:c+2]
            
            # Calculate the average of the 3x3 neighborhood
            average_value = np.mean(neighbors)
            
            # Assign the average to the corresponding pixel in the new image
            blurred_image[r, c] = average_value

    return blurred_image

# --- Example Usage ---
if __name__ == '__main__':
    # 1. Create a sample 2D grayscale image array
    # A 10x10 array with some distinct values
    original_image = np.zeros((10, 10))
    original_image[2:5, 2:5] = 255  # A bright square
    original_image[6:9, 6:9] = 100  # A gray square

    # 2. Apply the blur filter
    blurred_image = simple_blur(original_image)

    # 3. Display the original and blurred images
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    
    # Original image plot
    axes[0].imshow(original_image, cmap='gray', vmin=0, vmax=255)
    axes[0].set_title('Original Image')
    axes[0].axis('off')
    
    # Blurred image plot
    axes[1].imshow(blurred_image, cmap='gray', vmin=0, vmax=255)
    axes[1].set_title('Blurred Image')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

```
