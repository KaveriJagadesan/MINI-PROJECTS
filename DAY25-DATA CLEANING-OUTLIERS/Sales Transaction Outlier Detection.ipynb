```python
import pandas as pd
import numpy as np

# Sample sales transactions with some extreme values
data = {
    "transaction_id": range(1, 15),
    "total_amount": [120, 150, 130, 200, 180, 125, 170, 4000, 140, 135, 160, 12000, 155, 145]
}

df = pd.DataFrame(data)
print("Sales Transactions Dataset:")
print(df)
```

Visualize Total Amount with a Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(df['total_amount'], bins=10, kde=True)
plt.title("Histogram of Transaction Amounts")
plt.xlabel("Total Amount")
plt.ylabel("Frequency")
plt.show()
```

Calculate Z-Scores

```python
from scipy import stats

# Compute Z-scores
df['z_score'] = stats.zscore(df['total_amount'])

# Display dataset with Z-scores
print(df)
```

Identify Outliers Using Z-Score Threshold

```python
# Common threshold for outliers: |Z| > 3
threshold = 3
outliers = df[np.abs(df['z_score']) > threshold]

print("Identified Outliers:")
print(outliers)
```

Visualize Outliers

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['total_amount'])
plt.title("Box Plot of Total Amounts Showing Outliers")
plt.show()

```
