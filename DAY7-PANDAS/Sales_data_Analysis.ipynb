```python
import pandas as pd
```

# 1. Load Data

```python
data = {
    'Product': ['A', 'B', 'A', 'C', 'B'],
    'Quantity': [10, 5, 12, 8, 7],
    'Price': [20, 15, 20, 25, 15]
}
sales_df = pd.DataFrame(data)
```

# 2. Feature Engineering: Create 'Total Sales'

```python
sales_df['Total Sales'] = sales_df['Quantity'] * sales_df['Price']
```

# 3. Aggregation: Group by 'Product' and sum 'Total Sales

```python
product_sales = sales_df.groupby('Product')['Total Sales'].sum()
```

# 4. Output

```python
print("Total Sales per Product:\n")
print(product_sales)
```
