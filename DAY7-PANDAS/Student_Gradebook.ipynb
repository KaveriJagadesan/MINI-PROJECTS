1. Import Pandas and Create the DataFrame

```python
import pandas as pd

# Sample data: students and their grades
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92, 88, 82, 90],
    'Science': [90, 75, 95, 85, 80, 93],
    'English': [88, 82, 91, 90, 79, 89]
}

# Create DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

2. Calculate Average Score for Each Student

```python
# Group by 'Student' and take the mean of scores
student_averages = df.groupby('Student').mean(numeric_only=True)
print("\nAverage Score for Each Student:\n", student_averages)
```

3. Calculate Average Score for Each Subject

```python
# Calculate mean of each subject column
subject_averages = df[['Math', 'Science', 'English']].mean()
print("\nAverage Score for Each Subject:\n", subject_averages)
```
