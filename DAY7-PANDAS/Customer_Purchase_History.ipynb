```python
import pandas as pd
```

Sample Data: Customer purchase records

```python
data = {
    'CustomerID': [101, 102, 101, 103, 102, 101, 104, 103],
    'Product': ['Laptop', 'Phone', 'Mouse', 'Phone', 'Laptop', 'Keyboard', 'Mouse', 'Laptop'],
    'Amount': [80000, 30000, 1500, 30000, 85000, 2000, 1500, 85000]
}
```


Create DataFrame

```python
df = pd.DataFrame(data)
print("Customer Purchases DataFrame:\n", df, "\n")
```

Total spending for each customer

```python
total_spending = df.groupby('CustomerID')['Amount'].sum()
print("Total Spending per Customer:\n", total_spending, "\n")
```

Most popular product (most frequently purchased)

```python
most_popular_product = df['Product'].value_counts().idxmax()
print("Most Popular Product:", most_popular_product)
```
