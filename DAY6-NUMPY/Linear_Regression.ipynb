```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate Data
np.random.seed(42) 
X = np.random.rand(100, 1)
y = 2 * X.flatten() + 3 + np.random.randn(100) * 0.5
X_b = np.c_[np.ones((100, 1)), X]
# 2. Apply Normal Equation
beta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
intercept = beta[0]
slope = beta[1]
print(f"Calculated Intercept: {intercept:.4f}")
print(f"Calculated Slope: {slope:.4f}")
# 3. Make Predictions
predictions = X_b @ beta
# 4. Visualize
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Original Data Points', alpha=0.7)
plt.plot(X, predictions, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression using Normal Equation')
plt.legend()
plt.grid(True)
plt.show()
```
