```python
import pandas as pd
import matplotlib.pyplot as plt
```

1. Create DataFrame

```python
data = {
    'Category': ['Electronics', 'Clothing', 'Groceries', 'Furniture', 'Toys'],
    'Sales': [120000, 75000, 50000, 30000, 20000]
}

df = pd.DataFrame(data)
```

2. Aggregate Data

```python
category_sales = df.groupby('Category')['Sales'].sum()
```

 5. Subplots: side-by-side

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
```

 3. Bar Chart

```python
category_sales.plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title('Total Sales per Category')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Sales')
axes[0].tick_params(axis='x', rotation=45)
```

4. Pie Chart

```python
category_sales.plot(kind='pie', ax=axes[1], autopct='%1.1f%%', startangle=90, shadow=True)
axes[1].set_ylabel('')  # Remove default y-label for pie chart
axes[1].set_title('Market Share per Category')

plt.tight_layout()
plt.show()
```
