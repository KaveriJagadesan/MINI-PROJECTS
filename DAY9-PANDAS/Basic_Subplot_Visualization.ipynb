 Basic Subplot Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

 Create sample data

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)
```

Create a figure with 2 subplots (1 row, 2 columns)

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
```

Plot a line chart on the first subplot

```python
ax[0].plot(x, y, color='blue', linewidth=2, marker='o')
ax[0].set_title('Line Chart')
ax[0].set_xlabel('X values')
ax[0].set_ylabel('sin(X)')
```

Plot a scatter chart on the second subplot

```python
ax[1].scatter(x, y, color='red')
ax[1].set_title('Scatter Plot')
ax[1].set_xlabel('X values')
ax[1].set_ylabel('sin(X)')
```

Adjust layout and show the figure

```python
plt.tight_layout()
plt.show()
```
