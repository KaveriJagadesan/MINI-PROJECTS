```python
import numpy as np
import matplotlib.pyplot as plt
```

Define the function and its derivative

```python
def f(x):
    return x**2

def df(x):
    return 2*x
```

Initialize parameters

```python
# Starting point
x_start = 3.0  

# Hyperparameters
learning_rate = 0.1
epochs = 20

# Store the path for visualization
x_path = [x_start]
```

Gradient descent loop

```python
x = x_start
for _ in range(epochs):
    x = x - learning_rate * df(x)  # gradient descent update
    x_path.append(x)  # store the value
```

Visualization

```python
# Create points for the function
x_vals = np.linspace(-4, 4, 100)
y_vals = f(x_vals)

# Plot function
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label='f(x) = x^2')

# Plot gradient descent steps
plt.scatter(x_path, [f(xi) for xi in x_path], color='red', label='Gradient Descent Steps')
plt.plot(x_path, [f(xi) for xi in x_path], color='red', linestyle='--')

plt.title("Gradient Descent on f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
