```python
import numpy as np
import matplotlib.pyplot as plt

# Function and derivative
def f(x):
    return x**2

def df(x):
    return 2*x
```

Gradient descent function

```python
def gradient_descent_1d(x_start, learning_rate, epochs):
    x = x_start
    path = [x]
    for _ in range(epochs):
        x = x - learning_rate * df(x)
        path.append(x)
    return path
```

Run gradient descent with different learning rates

```python
x_start = 3.0
epochs = 20
learning_rates = [0.01, 0.1, 1.0]

paths = {}
for lr in learning_rates:
    paths[lr] = gradient_descent_1d(x_start, lr, epochs)
```

Plot the paths on the same graph

```python
# x values for the function
x_vals = np.linspace(-4, 4, 100)
y_vals = f(x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label='f(x) = x^2', color='black')

# Plot each gradient descent path
for lr, path in paths.items():
    plt.plot(path, [f(xi) for xi in path], marker='o', linestyle='--', label=f'lr = {lr}')

plt.title("Gradient Descent with Different Learning Rates")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
