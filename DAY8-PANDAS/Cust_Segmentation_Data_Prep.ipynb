```python
import pandas as pd
import numpy as np
```

Create a sample DataFrame

```python
data = {
    'CustomerID': [101, 102, 103, 104, 105],
    'Customer_Type': ['New', 'Regular', np.nan, 'Regular', 'New'],
    'Age': [25, np.nan, 35, 45, np.nan]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

Impute missing numerical data (Age) with the median

```python
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)
```

Handle categorical data

```python
# Fill missing 'Customer_Type' with 'Unknown'
df['Customer_Type'].fillna('Unknown', inplace=True)

# Perform one-hot encoding on 'Customer_Type'
df_encoded = pd.get_dummies(df, columns=['Customer_Type'])
```

Display the cleaned and encoded DataFrame

```python
print("\nCleaned and Encoded DataFrame:")
print(df_encoded)
```
