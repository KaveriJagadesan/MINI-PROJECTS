```python
import pandas as pd
import numpy as np
```

1.Create Two DataFrames

```python
sales_region_A = pd.DataFrame({
    'ProductID': [101, 102, 103, 104],
    'Sales': [2500, 3000, np.nan, 4000],
    'Region': ['A', 'A', 'A', 'A']
})

sales_region_B = pd.DataFrame({
    'product_id': [105, 106, 107, 108],
    'sales': [3500, np.nan, 4500, 5000],
    'Region': ['B', 'B', 'B', 'B']
})

print("Sales Region A:")
print(sales_region_A, "\n")
print("Sales Region B:")
print(sales_region_B, "\n")
```

2.Standardize and Merge

```python
# Rename columns in region B to match region A
sales_region_B.rename(columns={'product_id': 'ProductID', 'sales': 'Sales'}, inplace=True)

# Merge (stack) both DataFrames vertically
combined_sales = pd.concat([sales_region_A, sales_region_B], ignore_index=True)

print("Combined Sales Data (Before Cleaning):")
print(combined_sales, "\n")
```

3. Clean the Combined Data

```python
# Check for missing values
print("Missing Values Before Cleaning:")
print(combined_sales.isnull().sum(), "\n")

# Fill missing sales with the average of the 'Sales' column
combined_sales['Sales'].fillna(combined_sales['Sales'].mean(), inplace=True)

# Drop rows where ProductID is missing
combined_sales.dropna(subset=['ProductID'], inplace=True)

print("Combined Sales Data (After Cleaning):")
print(combined_sales, "\n")

# Check again for missing values
print("Missing Values After Cleaning:")
print(combined_sales.isnull().sum())
```
