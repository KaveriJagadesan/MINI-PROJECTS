```python
import pandas as pd
```

1. Create Employee Details Dataset

```python
employees = pd.DataFrame({
    'employee_id': [101, 102, 103, 104, 105],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'department': ['HR', 'Finance', 'IT', 'Finance', 'HR']
})
```

 2. Create Performance Review Dataset

```python
performance = pd.DataFrame({
    'employee_id': [101, 102, 103, 104, 105],
    'performance_score': [88, 92, 79, 85, 95]
})
```

3. Merge datasets on 'employee_id'

```python
merged_df = pd.merge(employees, performance, on='employee_id')
```

4. Calculate average performance per department

```python
avg_performance = merged_df.groupby('department')['performance_score'].mean().reset_index()
```

5. Display the result

```python
print("Merged Dataset:")
print(merged_df, "\n")
print("Average Performance Score per Department:")
print(avg_performance)
```
