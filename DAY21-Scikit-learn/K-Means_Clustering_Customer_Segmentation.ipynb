```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

1. Generate synthetic customer data

```python
X, y_true = make_blobs(
    n_samples=300,
    centers=3,
    cluster_std=1.0,
    random_state=42
)
```

2. Instantiate and fit the KMeans model

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

3. Get cluster labels and centers

```python
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

 4. Visualize the clusters

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50)
plt.scatter(centers[:, 0], centers[:, 1], marker='x', s=200, linewidths=3)
plt.title("Customer Segmentation using K-Means")
plt.xlabel("Feature 1 (e.g., Annual Spending)")
plt.ylabel("Feature 2 (e.g., Visit Frequency)")
plt.show()
```
