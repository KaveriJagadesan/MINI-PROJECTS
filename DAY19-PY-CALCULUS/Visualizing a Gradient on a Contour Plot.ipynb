```python
import numpy as np
import matplotlib.pyplot as plt
```

Define the function and its gradient

```python
def f(x, y):
    return x**2 + y**2

def grad_f(x, y):
    return np.array([2*x, 2*y])
```

Create a grid for plotting

```python
# Define grid
x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

Compute gradient vectors at each point

```python
U = 2 * X  # df/dx
V = 2 * Y  # df/dy
```

Plot contour and gradient vectors

```python
plt.figure(figsize=(8, 6))

# Contour plot
contours = plt.contour(X, Y, Z, levels=15, cmap='viridis')
plt.clabel(contours, inline=True, fontsize=8)

# Quiver plot (gradient vectors)
plt.quiver(X, Y, U, V, color='red')

plt.title("Contour and Gradient Field of f(x, y) = x^2 + y^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
